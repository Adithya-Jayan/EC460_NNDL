##### 1. (a).  Write python code for plotting the following activation functions for the input x in the range of -20 to 20. 
>- Sigmoid 
>- TanH 
>- Softsign 
>- ReLU 
>- Softmax

```python
##Defining required functions

def plot_sigmoid(Input,plot=1):
    output = Input.copy()
    for i in range(len(Input)):
        output[i] = np.exp(Input[i])/(np.exp(Input[i])+1)
    if(plot):
        plt.figure()
        plt.title("Sigmoid plot")
        plt.plot(Input,output)
    return(output)
    
def plot_tanh(Input,plot=1):
    output = Input.copy()
    for i in range(len(Input)):
        output[i] = (np.exp(Input[i])-np.exp(-1*Input[i]))/(np.exp(Input[i])+np.exp(-1*Input[i]))
    if(plot):
        plt.figure()
        plt.title("Tanh plot")
        plt.plot(Input,output)
    return(output)
    
def plot_softsign(Input,plot=1):
    output = Input.copy()
    for i in range(len(Input)):
        output[i] = Input[i]/(1+np.abs(Input[i]))
    if(plot):
        plt.figure()
        plt.title("Softsign plot")
        plt.plot(Input,output)
    return(output)
    
def plot_relu(Input,plot=1):
    output = Input.copy()
    for i in range(len(Input)):
        output[i] = max(0,Input[i])
    if(plot):
        plt.figure()
        plt.title("ReLU plot")
        plt.plot(Input,output)
    return(output)
    
def plot_softmax(Input,plot=1):
    output = Input.copy()
    denominator = sum(np.exp(Input))
    for i in range(len(Input)):
        output[i] = np.exp(Input[i])/denominator
    if(plot):
        plt.figure()
        plt.title("Softmax plot")
        plt.plot(Input,output)
    return(output)
```

```python
#Define 100 points in range [-20 to 20]
inputs = np.linspace(-20,20,100)

#Plotting the activation functions for given input range 
plot_sigmoid(inputs);
plot_tanh(inputs);
plot_softsign(inputs);
plot_relu(inputs);
plot_softmax(inputs);
```

##### 1. (b). Write python code for the derivative of the following activation functions and their plots for the input x in the range of -20 to 20.
>- Sigmoid 
>- TanH 
>- Softsign 
>- ReLU 

```python
##Defining derivative function

def derivative(function,Inputs,name):
    outputs = np.zeros(len(Inputs)-1)
    for i in range(len(Inputs)-1):
        outputs[i] = (function[i+1]-function[i])/(inputs[i+1]-inputs[i])
    plt.figure()
    plt.title(name + " derivative plot")
    plt.plot(Inputs[:-1],outputs)
```

```python
#Defining 100 inputs in given range [-20 to 20]
inputs = np.linspace(-20,20,100)

#Plotting derivatives of given functions
derivative(plot_sigmoid(inputs,plot=0),inputs,"sigmoid")
derivative(plot_tanh(inputs,plot=0),inputs,"tanh")
derivative(plot_softsign(inputs,plot=0),inputs,"softsign")
derivative(plot_relu(inputs,plot=0),inputs,"relu")
```

##### 3. Write python code to compute the following Regression loss value for the given true output and predicted output by network:

`# y_true = [11, 20,19,17,10,24,23] ## Target or actual Value `

`# y_pred =[12,18,19.5,18,9,23,24] ## Predicted value by ANN model `
> - MSE 
> - MAE 
> - MBE 
> - Huber Loss 
> - Epsilon Hinge Loss 
> - Square Epsilon Hinge Loss 
> - verify each loss value by analytical method 

```python
## Defining required functions

def MSE(y_true,y_pred):
    out = np.mean(np.square(y_true-y_pred))
    print("Mean squared error is {}".format(out))
    return out
    
def MAE(y_true,y_pred):
    out = np.mean(np.abs(y_true-y_pred))
    print("Mean absolute error is {}".format(out))
    return out
    
def MBE(y_true,y_pred):
    out = np.mean(y_pred-y_true)
    print("Mean bias error is {}".format(out))
    return out
    
def HuberLoss(y_true,y_pred,delta):
    diff = y_true-y_pred
    out = np.zeros_like(diff)
    for i in range(len(diff)):
        if(np.abs(diff[i])<delta):
            out[i] = 0.5*(diff[i]**2)
        else:
            out[i] = delta*diff[i] - 0.5*(delta**2)
    print("Huber Loss' for a delta value of {} is {}".format(delta,out))
    return out
    
def EpsilonHingeLoss(y_true,y_pred,epsilon):
    out = np.abs(y_pred-y_true) - epsilon
    out[out < 0] = 0
    print("Epsilon Hinge Loss' for epsilon value of {} is {}".format(epsilon,out))
    return out
    
def SquareEpsilonHingeLoss(y_true,y_pred,epsilon):
    out = 0.5 * (np.square(y_pred-y_true) - (epsilon**2))
    out[out < 0] = 0
    print("Square epsilon Hinge Loss' for epsilon value of {} is {}".format(epsilon,out))
    return out
    
```

```python
# Defining given truth and predicted values
y_true = np.array([11, 20,19,17,10,24,23]) ## Target or actual Value
y_pred = np.array([12,18,19.5,18,9,23,24]) ## Predicted value by ANN model

#Calculating requred loss functions
MSE(y_true,y_pred);
MAE(y_true,y_pred);
MBE(y_true,y_pred);
HuberLoss(y_true,y_pred,0.2);
EpsilonHingeLoss(y_true,y_pred,0.2);
SquareEpsilonHingeLoss(y_true,y_pred,1);
```

```python
## Defining required functions

def predict(X):
    return(0.7*X-0.1)
```

```python
# Regression to estimate coefficients y = a_1*x + a_0 
def estimate_coeff(x,y):
  #number
  n = np.size(x)

  #mean of x and y
  m_x = np.mean(x)
  m_y = np.mean(y)

  #calculating x_coefficient a_0
  num = np.sum(x*y)-n*m_x*m_y
  den = np.sum(x*x)-n*m_x*m_x

  #coefficients
  a_1 = num/den
  a_0 = m_y-a_1*m_x

  return(a_0,a_1)
```
